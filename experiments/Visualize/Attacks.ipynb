```python
def read_json(path):
  with open(path, 'r', encoding='utf-8') as curr_file:
    data = json.load(curr_file)
  return data
```

```python
def get_dict(data):
  dct = {
      'model':data['log']['model'],
      'task':data['log']['task'],
      'acc_train':data['log']['accuracy_Train'],
      'acc_test':data['log']['accuracy_Test'],
      'poisoning_rate':data['poisoning_rate']
  }
  return dct
```

```python
all_dcts = dcts1 + dcts2 + dcts3 + dcts4
```

```python
df = pd.DataFrame(all_dcts) 
```

```python
df
```

```python
df['model'] = df['model'].apply(lambda x: 'LR' if x == 'Logistic Regression' else x)
```

```python
df
```

```python
df['task'].unique()
```

```python
def changer(txt):
  if 'Attack' not in txt:
    return txt
  else:
    if 'Influe' in txt or 'Influence' in txt:
      return 'Influence'
    elif 'LabelFlip' in txt:
      return 'LabelFlip'
    elif 'Backdoor' in txt:
      return 'Backdoor'
```

```python
df['attack'] = df['task'].apply(lambda x: changer(x))
```

```python
df
```

```python
def changer(txt):
  if 'tf-idf' in txt:
    return 'IMDB'
  elif '0-1' in txt:
    return 'MNIST (0-1)'
  else:
    return txt
```

```python
df['dataset'] = df['task'].apply(lambda x: changer(x))
```
